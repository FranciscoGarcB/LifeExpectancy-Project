```python
import pandas as pd
df = pd.read_csv("datasets/LifeExpectancy.csv")
df.head(4)
```

```python
# Columnas del dataset
df.columns
```

```python
# Years included on the dataset
df["Year"].unique()
```

```python
# Countries included on the dataset
print(f'Unique countries: {df["Country"].nunique()}')
for country in df["Country"].unique():
    print(country)
```

```python
# Change countries name to coincide with the GeoJSON data
df['Country'] = df['Country'].replace('Bolivia (Plurinational State of)', 'Bolivia')
df['Country'] = df['Country'].replace('Venezuela (Bolivarian Republic of)', 'Venezuela')
df['Country'] = df['Country'].replace('Democratic Republic of the Congo', 'Dem. Rep. Congo')
df['Country'] = df['Country'].replace('Russian Federation', 'Russia')
df['Country'] = df['Country'].replace('Dominican Republic', 'Dominican Rep.')
df['Country'] = df['Country'].replace('United Republic of Tanzania', 'Tanzania')
df['Country'] = df['Country'].replace('South Sudan', 'S. Sudan')
df['Country'] = df['Country'].replace('Central African Republic', 'Central African Rep.')
df['Country'] = df['Country'].replace('Iran (Islamic Republic of)', 'Iran')
df['Country'] = df['Country'].replace('Syrian Arab Republic', 'Syria')
df['Country'] = df['Country'].replace('Bosnia and Herzegovina', 'Bosnia and Herz.')
df['Country'] = df['Country'].replace('Republic of Moldova', 'Moldova')
df['Country'] = df['Country'].replace('Republic of Moldova', 'Moldova')
df['Country'] = df['Country'].replace('The former Yugoslav republic of Macedonia', 'North Macedonia')
df['Country'] = df['Country'].replace('United Kingdom of Great Britain and Northern Ireland', 'United Kingdom')
df['Country'] = df['Country'].replace('Equatorial Guinea', 'Eq. Guinea')
df['Country'] = df['Country'].replace("Lao People's Democratic Republic", 'Laos')
df['Country'] = df['Country'].replace('Viet Nam', 'Vietnam')
df['Country'] = df['Country'].replace('Republic of Korea', 'South Korea')
df['Country'] = df['Country'].replace("Democratic People's Republic of Korea", 'North Korea')
df['Country'] = df['Country'].replace('Solomon Islands', 'Solomon Is.')
df['Country'] = df['Country'].replace('Micronesia (Federated States of)', 'Micronesia')
df['Country'] = df['Country'].replace('United States of America', 'United States')
```

```python
# Save updated files
df.to_csv('datasets/LifeExpectancyUpdated.csv', index=False)
```

## Short names

```python
# For shorter names (uses less margins on barchart)
df1 = df.copy()
df1 = df1[['Year', 'Life expectancy ', 'Status', 'GDP', 'Adult Mortality', 'infant deaths', ' HIV/AIDS']]
df1['Country'] = df['Country'].replace('Saint Vincent and the Grenadines', 'St. Vincent & the Grenadines')
df1 = df1.dropna(subset=['Year', 'Life expectancy '])
df1.to_csv('datasets/LifeExpectancyShort.csv', index=False)
```

## Developed vs developing

```python
df2 = df.copy()
print(f'Number of developed countries: {df["Country"][df["Status"]=="Developed"].nunique()}')
print(f'Number of developed countries: {df["Country"][df["Status"]=="Developing"].nunique()}')
df2.rename(columns={"Life expectancy ": "LifeExpectancy"}, inplace=True)
df2.to_csv("datasets/DevelopedLE.csv")
```
