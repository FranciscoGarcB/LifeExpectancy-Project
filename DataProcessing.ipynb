```python
# Save updated files
df.to_csv('datasets/LifeExpectancyUpdated.csv', index=False)
```

```python
import pycountry_convert as pc

# Función para obtener el continente de un país
def country_to_continent(country_name):
    try:
        # Obtener el código del país
        country_code = pc.country_name_to_country_alpha2(country_name)
        # Obtener el código del continente
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        # Convertir código de continente a nombre
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except:
        return None  # En caso de que el país no sea encontrado

# Crear la columna 'Continent'
df['Continent'] = df['Country'].apply(country_to_continent)

# Mostrar el DataFrame resultante
print(df)
```
